```python
from euroleague_api.EuroLeagueData import EuroLeagueData
from euroleague_api.boxscore_data import BoxScoreData
from euroleague_api.player_stats import PlayerStats

import pandas as pd
pd.set_option('future.no_silent_downcasting', True)
```

```python
previous_season = 2023
current_season = 2024
game_code = 1
competition_code = "E"

season_data = EuroLeagueData(competition_code)
df_season_data_previous = season_data.get_game_metadata_season(season=previous_season)
df_season_data_current = season_data.get_game_metadata_season(season=current_season)
```

```python
df_season_data_current.head()
```

```python
df_season_data = pd.concat([df_season_data_previous, df_season_data_current])
```

```python
# Combine the 'date' and 'time' columns into a single column
df_season_data['datetime'] = df_season_data['date'] + ' ' + df_season_data['time']

# Convert the combined column into a datetime object
df_season_data['datetime'] = pd.to_datetime(df_season_data['datetime'], format='%b %d, %Y %H:%M')

# drop date, time
df_season_data.drop(['date', 'time'], inplace=True, axis=1)
```

```python
df_season_data = df_season_data.sort_values(by='datetime')
```

```python
df_season_data.tail(10)
```

```python
boxscore = BoxScoreData(competition_code)

boxscore.get_boxscore_data(season=current_season,gamecode=1, boxscore_type='Stats')
```

```python
boxscore.get_player_boxscore_stats_data(season=current_season, gamecode=1).iloc[14]
```

```python
alba_pao = boxscore.get_player_boxscore_stats_data(season=current_season, gamecode=1)
```

```python
alba_pao.tail(15)
```

```python
df_players_prev_season = pd.read_csv('Players_season_2023_2024.csv', sep=';')
df_players_curr_season = pd.read_csv('Players_season_2024_2025.csv', sep=';')
```

```python
df_players_prev_season.head()
```

```python
# Define a regex pattern to match valid player codes (starting with P followed by 5 to 7 digits)
pattern = r"^P\d{6}$"

alba_pao['Player_ID'] = alba_pao['Player_ID'].str.strip()

# Filter rows where 'Player_ID' matches the pattern
alba_pao = alba_pao[alba_pao['Player_ID'].str.match(pattern)]

# Replace 'DNP' in the 'Minutes' column with '00:00'
alba_pao['Minutes'] = alba_pao['Minutes'].replace('DNP', '00:00')

alba_pao
```

```python
# Function to convert "LASTNAME, FIRSTNAME" to "F. Lastname"
def format_name(name):
    last, first = name.split(', ')
    return f"{first[0]}. {last.capitalize()}"
```

```python
alba_pao['PlayerName'] = alba_pao['Player'].apply(format_name)
```

```python
alba_pao
```

```python
# Step 1: Concatenate the two DataFrames
df_concat = pd.concat([df_players_prev_season, df_players_curr_season])

# Step 2: Remove duplicate players (keeping only the last occurrence)
df_concat = df_concat.drop_duplicates(subset=['Player'], keep='last')

# Step 3: Keep only 'Player' and 'Pos' columns
df_concat = df_concat[['Player', 'Pos']]

# Step 4: Rename 'Player' column to 'PlayerName' to match the first table
df_concat.rename(columns={'Player': 'PlayerName'}, inplace=True)

df_concat.head(10)
```

```python
# Perform a left merge on the 'PlayerName' column
merged_df = pd.merge(alba_pao, df_concat, on='PlayerName', how='left')
```
